#Solución analítica 
Problema 
Un paracaidista con una masa de $68.1 kg$ salta de un globo aeros­tático fijo. Aplique 
la ecuación (1.10) para calcular la velocidad antes de que se abra el paracaídas. 
Considere que el coeficiente de arrastre es igual a $12.5 kg/s.$ 
___
$$v(t)=\frac{gm}{t}(1-e^{-(c/m)t})$$



```
for x in range(0,12,2):
  print(round(((9.81*68.1)/12.5)*(1-e**((-12.5/68.1)*x)),2))
```
